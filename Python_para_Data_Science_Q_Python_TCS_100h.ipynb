## Cientista de Dados
Um cientista de dados deve ter os conhecimentos principalmente em machine learning, estatística, programação, matemática e banco de dados.

Como um cientista de dados trabalha?
- Realizar as perguntas corretas para entender o problema.
- Explorar e coletar os dados de alguma fonte.
- Extrair os dados, transformar os dados em um padrão.
- Limpar os dados para remover valores errados ou que podem interferir no resultado.
- Preencher lacunas para que a análise seja feita corretamente. As lacunas podem ser preenchidas com valores médios ou outro valor.
- Normalizar dados: os dados podem vir em diferentes escalas e isso deve ser padronizado. Por exemplo: em um registro pode ser encontrado o valor de 50 cm e em outro com o mesmo valor, porém registrado em 0,5m.
- Analisar os dados, encontrar padrões e realizar predições.
- Apresentar os resultados com conclusões e sugestões de maneira que outras pessoas entendam facilmente.



## Dados
Os dados são coleções de informações que podem se apresentar de forma estruturada ou não. Dados não estruturados devem ser organizados.
Por exemplo:
Pedro realizou uma pesquisa de satisfação de consumidores de uma loja. O método que foi utilizado foi a comunicação por mensagens instantâneas e e-mails. Pedro recebeu respostas de forma desestruturada, ou seja, ele deverá realizar a estruturação desses dados em uma planilha para obter os primeiros resultados.
Já Juliana realizou a mesma pesquisa, entretando ela enviou um link para os clientes, direcionando-os para um formulário eletrônico. A grande vantagem que ela obteve foi a de que o sistema de formulário eletrônico já entregava a ela uma planilha eletrônica com todos os dados estruturados.

Tipos de dados estuturados: listas, tabelas, bancos de dados...



## Python
Como trabalhar com Ciência de Dados utilizando Python?
É comum a utilização de algumas bibliotecas, como por exemplo:

- NumPy: é uma biblioteca matemática para trabalhar com vetores, álgebra linear, transformada de Fourier, etc.
- Pandas: é uma biblioteca para trabalhar com dados estruturados. Com ela podemos importar arquivos CSV, ARFF, XLSX, além de criar dataframes - que são tabelas e realizar a preparação de dados.
- Matplotlib: é uma biblioteca para plotar gráficos a partir de dados.
- SciPy: é uma biblioteca de álgebra linear 
- Scikit-learn: é uma biblioteca de machine learning que contém algoritmos de classificação, regressão e agrupamento. Alguns algoritmos que incluem na biblioteca são SVM, Random Forest, Gradient Boosting, K-means, DBSCAN, etc.
- River: é uma biblioteca de machine learning para trabalhar com fluxo contínuo de dados e big data.

##Machine Learning



###Classificação
Processo de buscar e encontrar um modelo para separar dados em classes categóricas. Os dados são classificados em diferentes classes dependendo dos parâmetros e então rótulos podem ser previstos para os dados.

Exemplos de classificação: 
- binária: sim e não, classe 1 e classe 2, cachorro ou gato
- multi-classe: time a, b ou c; cachorro, gato ou rato.

Algoritmos de classificação em ML:
- Regressão Logística
- KNN
- SVN
- Naïve Bayes
- Árvores de decisão
- Random Forest

###Regressão
A regressão busca encontrar um modelo ou função para distinguir dados em valores reais contínuos ao invés de classficar utilizando rótulos.
Um algoritmo de regressão busca entrar uma função mapeadora para uma entrada para obter uma saída contínua.

Exemplos:
- Previsão do tempo
- Valor de uma ação em bolsa de valores
- etc.

Algoritmos de Regressão:
- Regressão linear simples
- Regressão linear múltipla
- Regressão Polinomial
- Regressão de Vetor de Suporte
- Regressão de árvore de decisão
- Regressão de Random Forest

## Exemplo de Regressão



### Regressão Linear

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y, 'o')
plt.show()
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}") #quando x = 0, o valor é 5,63
print(f"slope: {model.coef_}")#o incremento a cada x+1
```

```python
y_pred = model.predict(x)
print(f"predicted response:\n{y_pred}")
```

```python
#Resultado igual, porém aplicando o resultado na reta.
y_pred = model.intercept_ + model.coef_ * x
print(f"predicted response:\n{y_pred}")
```

```python
plt.plot(x, y, 'o')
plt.plot(x, y_pred)
plt.show()
```

```python
x2 = [[65],[85]]
y_pred2 = model.predict(x2)
print(y_pred2)
```

```python
plt.plot(x, y, 'o')
plt.plot(x, y_pred)
plt.plot(x2,y_pred2)
plt.show()
```

### Regressão Polinomial

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([15, 11, 2, 8, 25, 32])
```

```python
plt.plot(x, y, 'o')
plt.show()
```

```python
transformer = PolynomialFeatures(degree=2, include_bias=False)
transformer.fit(x)
```

```python
x_ = transformer.transform(x)
x_ = PolynomialFeatures(degree=2, include_bias=False).fit_transform(x)
x_
```

```python
model = LinearRegression().fit(x_, y)
r_sq = model.score(x_, y)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}")
print(f"coefficients: {model.coef_}")
```

```python
y_pred = model.predict(x_)
y_pred
```

```python
plt.plot(x, y, 'o')
plt.plot(x,y_pred, 'x')
plt.show()
```

```python
y_pred = model.predict([[65, 4225]])
y_pred
```
